```python
import pandas as pd

class Pb:
    def __init__(self):
        self.records = []

    def read_from_csv(self, file_path):
        df = pd.read_csv(file_path)
        self.records = df.to_dict(orient='records')

class PbSQLParser:
    def __init__(self, phonerecords):
        self.phonerecords = phonerecords

    def execute_query(self, query):
        query_parts = query.split()

        if query_parts[0].upper() == 'SELECT':
            return self._execute_select_query(query_parts)

    def _execute_select_query(self, query_parts):
        if 'FROM' in query_parts and query_parts[query_parts.index('FROM') + 1].lower() == 'sql':
            if '*' in query_parts:
                return self.phonerecords.records[:10]
            elif 'WHERE' in query_parts:
                condition_field = query_parts[query_parts.index('WHERE') + 1]
                condition_value = query_parts[query_parts.index('=') + 1].strip("';")
                return [record for record in self.phonerecords.records if record.get(condition_field) == condition_value]

sql = Pb()
sql.read_from_csv('sql.csv')  # Assuming 'sql.csv' is your CSV file
parser = PbSQLParser(sql)
result = parser.execute_query("SELECT * FROM sql WHERE name='Doe'")
print(result)
```
